```python
#最大の数値を出力
a = 8.9
b = 10.7
c = 1.1
if a <= b:
  if b <= c:
    print(c)
  else :
    print(b)
else :
  if a <= c:
    print(c)
  else:
    print(a)
```

```python
#繰り返し(数値の繰り返し)
for n in range(10):
  print(3 * n + 7)
```

```python
#リストの繰り返し
l = ['a', 'b', 'c', 'd', 'e']
for x in l: #lから一つずつ要素を取り出し、xに代入
  print(x)
```

```python
#リストの大きさを利用
for i in range(len(l)):
  print(l[i])
```

```python
#リストへ要素の追加
l.append('f')
l
```

```python
#enumerateの利用
word = ['A', 'B', 'C']
for i, val in enumerate(word):
  print('index : {0}, value = {1}'.format(i, val))
```

```python
#zipの利用
word1 = ['A', 'B', 'C']
word2 = ['X', 'Y', 'Z']
for i, j in zip(word1, word2):
  print('word1 : {0}, word2 = {1}'.format(i, j))
```

```python
#リストの内包利用
l = [i**2 for i in range(10)]
print(l)
```

```python
#リストの内包表記(条件付き)
even = [i for i in range(0, 21) if i % 2 == 0]
print(even)
```
